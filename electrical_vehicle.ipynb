```python
import matplotlib.pyplot as plt

data['Base MSRP'].hist(bins=30)
plt.title('Distribution of Base MSRP')
plt.xlabel('Base MSRP')
plt.ylabel('Frequency')
plt.show()
```

```python
print(data.columns)
```

```python
make_counts = data['Make'].value_counts().reset_index()
make_counts.columns = ['Make','Count']
fig = px.bar(make_counts,
             x='Make',
             y='Count',
             title='Number of Electric Vehicles by Make',
             labels={'Make':'Vehicle Make','Count':'Number of Vehicles'},
             color='Count',
             color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```

```python
fig = px.box(data,
             x='Make',
             y='Base MSRP',
             title='Box Plot of Base MSRP by Electric Vehicle Make',
             labels={'Make': 'vehicle Make','Base MSRP': 'Base MSRP ($)'},
             color='Make')
fig.show()
```

```python
fig = px.bar(data,
             x='Base MSRP',
             y='Make',
             color='Make',
             animation_frame='Model Year',
             range_x=[0,data['Base MSRP'].max()],
             title='Racing Bar Plot pf EV Makes Over Years')
fig.show()
```
